```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
```

```python
attack_test = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/attack_test_8020.csv')
attack_train = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/attack_train_8020.csv')
normal_test = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/normal_test_8020.csv')
normal_train = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/normal_train_8020.csv')
```

```python
train_df = pd.merge(attack_train, normal_train, how='outer')
```

```python
test_df = pd.merge(attack_test, normal_test, how='outer')
```

```python
train_df['attack_type'] = np.where(train_df.attack_type == 'attack', 0, train_df.attack_type)
train_df['attack_type'] = np.where(train_df.attack_type == 'normal', 1, train_df.attack_type)
test_df['attack_type'] = np.where(test_df.attack_type == 'attack', 0, test_df.attack_type)
test_df['attack_type'] = np.where(test_df.attack_type == 'normal', 1, test_df.attack_type)
```

```python
train_df
```

```python
train_df.attack_type = train_df.attack_type.astype(int)
test_df.attack_type = test_df.attack_type.astype(int)
```

```python
train_df = shuffle(train_df, random_state = 7)
test_df = shuffle(test_df, random_state = 7)
```

```python
y_train = train_df['attack_type']
y_test = test_df['attack_type']

train_df = train_df.drop('attack_type', axis = 1)
test_df = test_df.drop('attack_type', axis = 1)
```

```python
rf = RandomForestClassifier(n_estimators = 20, random_state = 42)
rf.fit(train_df, y_train)
```

```python
preds = rf.predict(test_df)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
CLASS = ['Attack', 'Normal']
```

```python
check_element = 5
result = np.where(preds == check_element)

print(f'First n predictions:\n\n {(preds[:10])}\n')
print(f'Category classification of element at index {check_element}: {CLASS[np.argmax(preds[check_element])]}')
```

```python
print(f'{confusion_matrix(y_test, preds)}\n\n{round(accuracy_score(y_test, preds), 4)*100}\n\n{classification_report(y_test, preds, target_names = CLASS)}')
```
